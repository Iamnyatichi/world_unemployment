```python
# Importing libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import leafmap 
```

```python
# Loading data and reading
unemployment=pd.read_csv('https://raw.githubusercontent.com/abdullahkhanjaan/World-unemployment-Data-analysis/refs/heads/main/data.csv', delimiter=',')
world_geo=gpd.read_file('https://raw.githubusercontent.com/EBISYS/Mapping-Dam-Levels/refs/heads/master/world-countries.json')
unemployment
```

```python
world_geo
```

```python
#extracting 2021 data
unemployment=unemployment[['Country Name','2021']]
unemployment
```

```python
#Replacing missing countries in the data
#replace names
replace_names={
    'Congo, Dem. Rep.':'Democratic Republic of the Congo',
    "Cote d'Ivoire": "Ivory Coast",
    'Yemen, Rep.':'Yemen',
    'Bahamas, The':'The Bahamas',
    'Iran, Islamic Rep.':'Iran',
    'North Macedonia': 'Macedonia',
    'Venezuela, RB':'Venezuela',
    "Korea, Dem. People's Rep.":"North Korea", 
    'United States':'United States of America',
    'Egypt, Arab Rep.':'Egypt',
    'Russian Federation':'Russia',
    'Kyrgyz Republic':'Kyrgyzstan',
    'Tanzania':'United Republic of Tanzania',
    'Eswatini':'Swaziland',
    'Serbia':'Republic of Serbia',
    'Gambia, The':'Gambia',
    'Lao PDR':'Laos',
    'Guinea-Bissau':'Guinea Bissau',
    'Korea, Rep.':'North Korea',
    'Syrian Arab Republic':'Syria',
    'West Bank and Gaza':'West Bank',
    'Brunei Darussalam':'Brunei',
    'Timor-Leste':'East Timor',
    'Korea, Rep.':'South Korea',
    'Congo, Rep.':'Republic of the Congo',
    'Slovak Republic':'Slovakia',
    'Turkiye':'Turkey'
    
    }
unemployment['Country Name']=unemployment['Country Name'].replace(replace_names)
```

```python
# ensuring matching columns
world_geo=world_geo.rename(columns={'name':'Country Name'})


#Check data completeness in both df and geodataframe
missing_countries=set(world_geo['Country Name'])-set(unemployment['Country Name'])
for country in missing_countries:
    print(f'The country, {country} is not in the unemployment data')
```

```python
#Check data completeness in both df and geodataframe
missing_countries=set(unemployment['Country Name'])-set(world_geo['Country Name'])
for country in missing_countries:
    print(f'The country, {country} is not in the world_geo data')
```

```python
# merging the two datasets
unemployment_data=world_geo.merge(unemployment, on='Country Name')
unemployment_data
```

```python
#Visualize
m=leafmap.Map()
m.add_data(unemployment_data, 
            column='2021',
           layer_name='World Unemployment Rates',
           legend_title='unemployment Rate',
           cmap='Spectral_r',
           opacity=0.5
                     )


#export to html
m.to_html('unemployment_rate.html')

#visualize
m
```
